#### 1. Write Python code and use MapReduct to count occurrences of each word in the first text file (file.txt). How many times each word is repeated

```python
# replaced /n with space in the text file
with open("file1.txt", "r") as file:
    newtext = ""
    for line in file:
        line_strip = line.replace('\n', " ")
        newtext += line_strip
    print(newtext)  
```

```python
## Removed numbers and punctuations from the text file in order to count the words only
```

```python
text_without_numbers = ''.join([i for i in newtext if not i.isdigit()])
print(text_without_numbers[0:100])
```

```python
import re
text_without_punc = re.sub(r'[^\w\s]','',text_without_numbers)
print(text_without_punc[0:100])
```

```python
## splitted the text into list of words
```

```python
word_text = text_without_punc.split()  
print(word_text[0:10])
```

```python
## mapping the words
```

```python
words = []
for item in word_text:
    temp = item +' ' + str(1)
    words.append(temp)
print(words[0:10])
```

```python
words.sort()
print(words[0:10])
```

```python
## Reducing the mapped words
```

```python
reduced_words = list(dict.fromkeys(words))
for item in reduced_words:
    count = words.count(item)
    print(item.split(' ')[0] + ":" + str(count))
```
